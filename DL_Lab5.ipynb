```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense
from tensorflow.keras.utils import to_categorical
```

```python
text = """
Artificial intelligence is transforming the world.
Machine learning is a part of artificial intelligence.
Deep learning is a powerful technique in machine learning.
Artificial intelligence helps machines learn from data.
Neural networks are inspired by the human brain.
Data science and artificial intelligence work together.
Machine learning models improve with more data.
Artificial intelligence is widely used in industries.
""" * 20

print("Text Loaded Successfully")
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

total_words = len(tokenizer.word_index) + 1

print("Total Vocabulary Size:", total_words)

input_sequences = []

for line in text.split('\n'):

    token_list = tokenizer.texts_to_sequences([line])[0]

    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

print("Total Training Sequences:", len(input_sequences))
```

```python
max_sequence_len = max([len(x) for x in input_sequences])

input_sequences = np.array(
    pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')
)

print("Max Sequence Length:", max_sequence_len)
```

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

y = to_categorical(y, num_classes=total_words)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
rnn_model = Sequential()

rnn_model.add(Embedding(total_words, 50, input_length=max_sequence_len-1))
rnn_model.add(SimpleRNN(100))
rnn_model.add(Dense(total_words, activation='softmax'))

rnn_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

print("RNN Model Summary")
rnn_model.summary()
```

```python
rnn_model.fit(X, y, epochs=100, verbose=1)
```

```python
lstm_model = Sequential()

lstm_model.add(Embedding(total_words, 50, input_length=max_sequence_len-1))
lstm_model.add(LSTM(150))
lstm_model.add(Dense(total_words, activation='softmax'))

lstm_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

print("LSTM Model Summary")
lstm_model.summary()
```

```python
lstm_model.fit(X, y, epochs=100, verbose=1)
```

```python
def generate_text(seed_text, next_words, model):

    for _ in range(next_words):

        token_list = tokenizer.texts_to_sequences([seed_text])[0]

        token_list = pad_sequences(
            [token_list],
            maxlen=max_sequence_len-1,
            padding='pre'
        )

        predicted = np.argmax(model.predict(token_list), axis=-1)

        output_word = ""

        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break

        seed_text += " " + output_word

    return seed_text
```

```python
seed_text = "artificial"

print("\nGenerated Text using RNN:")
print(generate_text(seed_text, 5, rnn_model))

print("\nGenerated Text using LSTM:")
print(generate_text(seed_text, 5, lstm_model))
```
